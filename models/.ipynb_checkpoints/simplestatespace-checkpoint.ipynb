```python
# choose det_order: -1=no deterministic terms, 0=constant, 1=constant+trend
det_order = 0

# choose k_ar_diff = number of lagged differences (i.e., underlying VAR lags - 1)
# You can choose it by information criteria from a VAR on levels. Let's select via VAR AIC:
maxlags = 18
var_model = VAR(df)
ic = var_model.select_order(maxlags)
print(ic.summary())   # pick lag order for VAR in levels (p). Then k_ar_diff = p - 1

p = ic.selected_orders['aic'] or ic.selected_orders['bic'] or ic.selected_orders['hqic']
if p is None:
    p = 2
k_ar_diff = max(1, p - 1)

jres = coint_johansen(df.values, det_order, k_ar_diff)
print("Trace statistics:", jres.lr1)
print("Critical values (90/95/99):", jres.cvt)
# Decide cointegration rank r by comparing trace stats to critical values.
```

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.vector_ar.vecm import VECM
import time

def vecm_bootstrap_intervals(df,
                             k_ar_diff=2,
                             coint_rank=1,
                             deterministic="ci",
                             n_months=12,
                             B=500,
                             seed=12345):
    """
    Fit VECM on df and produce bootstrap forecast intervals.
    Robust handling of res.gamma shapes.
    Returns: point_df, mean_df, lower_df, upper_df
    """
    # Fit VECM
    vecm = VECM(df, k_ar_diff=k_ar_diff, coint_rank=coint_rank, deterministic=deterministic)
    res = vecm.fit()

    # Point forecast (levels)
    point_forecast = res.predict(steps=n_months)

    # Forecast index: start at next month after last observed
    last_date = pd.to_datetime(df.index[-1])
    # ensure monthly period start alignment: add one month
    first_forecast_month = last_date + pd.offsets.MonthEnd(1)
    forecast_index = pd.date_range(start=first_forecast_month, periods=n_months, freq='M')

    point_df = pd.DataFrame(point_forecast, index=forecast_index, columns=df.columns)

    # Extract alpha, beta
    alpha = getattr(res, "alpha", None)
    beta = getattr(res, "beta", None)
    if alpha is None or beta is None:
        raise RuntimeError("res.alpha or res.beta missing from VECMResults; cannot proceed.")

    n = alpha.shape[0]

    # === Robust extraction/reshaping of Gamma ===
    Gamma_raw = getattr(res, "gamma", None)
    if Gamma_raw is None:
        Gamma = np.zeros((k_ar_diff, n, n))
    else:
        Gamma_arr = np.array(Gamma_raw)
        # Case 1: already (k, n, n)
        if Gamma_arr.ndim == 3 and Gamma_arr.shape[1] == n and Gamma_arr.shape[2] == n:
            layers = Gamma_arr.shape[0]
            Gamma = np.zeros((k_ar_diff, n, n))
            Gamma[:layers] = Gamma_arr[:k_ar_diff]
        # Case 2: returned as (n, n)  -> single lag matrix
        elif Gamma_arr.ndim == 2 and Gamma_arr.shape == (n, n):
            Gamma = np.zeros((k_ar_diff, n, n))
            Gamma[0] = Gamma_arr
        # Case 3: returned as (n, n*k) horizontally concatenated
        elif Gamma_arr.ndim == 2 and Gamma_arr.shape[0] == n and Gamma_arr.shape[1] == n * k_ar_diff:
            Gamma = np.zeros((k_ar_diff, n, n))
            for i in range(k_ar_diff):
                Gamma[i] = Gamma_arr[:, i*n:(i+1)*n]
        # Case 4: returned as (k*n, n) vertically stacked
        elif Gamma_arr.ndim == 2 and Gamma_arr.shape[0] == k_ar_diff * n and Gamma_arr.shape[1] == n:
            Gamma = np.zeros((k_ar_diff, n, n))
            for i in range(k_ar_diff):
                Gamma[i] = Gamma_arr[i*n:(i+1)*n, :]
        # Case 5: unexpected but maybe (n, m) where m divisible by n
        elif Gamma_arr.ndim == 2 and Gamma_arr.shape[0] == n and (Gamma_arr.shape[1] % n == 0):
            potential_k = Gamma_arr.shape[1] // n
            use_k = min(potential_k, k_ar_diff)
            Gamma = np.zeros((k_ar_diff, n, n))
            for i in range(use_k):
                Gamma[i] = Gamma_arr[:, i*n:(i+1)*n]
            # warn user (print) that we truncated/padded
            print(f"Warning: res.gamma was shape {Gamma_arr.shape}. Interpreted as {use_k} layers and padded to {k_ar_diff}.")
        else:
            raise ValueError(f"Unexpected res.gamma shape {Gamma_arr.shape} for n={n} and k_ar_diff={k_ar_diff}. "
                             "Please paste the shape if you need further handling.")

    # === constant in cointegration space (robust) ===
    const_coint = None
    for attr in ("det_coef_coint", "det_coef", "det_coef_dtr", "deterministic_coef", "deterministic_coeff"):
        if hasattr(res, attr):
            c = getattr(res, attr)
            try:
                const_coint = np.asarray(c).reshape(-1)
            except:
                const_coint = None
            if const_coint is not None:
                break
    if const_coint is None:
        const_coint = np.zeros(n)

    # Residuals
    resid = res.resid
    if resid.ndim != 2:
        raise RuntimeError(f"res.resid has unexpected shape {resid.shape}")

    T, n_vars = resid.shape
    if n_vars != n:
        # defensive check
        raise RuntimeError(f"Number of variables mismatch: alpha.shape[0]={n}, resid.shape[1]={n_vars}")

    # initial states
    last_levels = df.values[-1].astype(float).copy()
    diffs = df.diff().dropna()
    # collect last k_ar_diff diffs (pad with zeros if not enough history)
    lagged_diffs = []
    for i in range(1, k_ar_diff + 1):
        if len(diffs) >= i:
            lagged_diffs.append(diffs.values[-i].astype(float))
        else:
            lagged_diffs.append(np.zeros(n))
    lagged_diffs = np.array(lagged_diffs)

    # simulation routine
    def simulate_path(eps_seq):
        X_prev = last_levels.copy()
        diffs_buffer = lagged_diffs.copy()
        steps = eps_seq.shape[0]
        sim_levels = np.zeros((steps, n))
        for t in range(steps):
            ec_term = alpha @ (beta.T @ X_prev)         # shape (n,)
            short_run = np.zeros(n)
            for j in range(k_ar_diff):
                short_run += Gamma[j].dot(diffs_buffer[j])
            const_term = alpha.dot(const_coint)
            delta = ec_term + short_run + const_term + eps_seq[t]
            X_next = X_prev + delta
            sim_levels[t] = X_next
            # update diffs buffer
            if k_ar_diff > 1:
                diffs_buffer[1:] = diffs_buffer[:-1]
            if k_ar_diff >= 1:
                diffs_buffer[0] = delta
            X_prev = X_next
        return sim_levels

    # bootstrap
    rng = np.random.default_rng(seed)
    all_forecasts = np.zeros((B, n, n))  # placeholder; will reshape after we know dims
    # allocate properly
    all_forecasts = np.zeros((B, n, n))  # temp, will replace next line
    all_forecasts = np.zeros((B, n_months, n))

    start = time.time()
    for b in range(B):
        idx = rng.integers(low=0, high=T, size=n_months)
        eps_seq = resid[idx, :]
        sim = simulate_path(eps_seq)
        all_forecasts[b] = sim
        if (b+1) % max(1, B//10) == 0:
            print(f"Bootstrap {b+1}/{B} done, elapsed {time.time()-start:.1f}s")

    mean_forecast = all_forecasts.mean(axis=0)
    lower_forecast = np.percentile(all_forecasts, 2.5, axis=0)
    upper_forecast = np.percentile(all_forecasts, 97.5, axis=0)

    mean_df = pd.DataFrame(mean_forecast, index=forecast_index, columns=df.columns)
    lower_df = pd.DataFrame(lower_forecast, index=forecast_index, columns=df.columns)
    upper_df = pd.DataFrame(upper_forecast, index=forecast_index, columns=df.columns)

    point_df.index = forecast_index

    return point_df, mean_df, lower_df, upper_df
```

```python
point_df, mean_df, lower_df, upper_df = vecm_bootstrap_intervals(
    df,
    k_ar_diff=6,
    coint_rank=1,
    deterministic="ci",
    n_months=12,
    B=500,
    seed=12345
)
```

```python
# --- Define colors for consistency ---
colors = {
    'Ap': '#FF0000',        # Red
    'Hoyre': '#0000FF',     # Blue
    'Frp': '#00008B',       # Dark Blue
    'SV': '#FF6347',        # Light Red (Tomato)
    'SP': '#006400',        # Dark Green
    'KrF': '#FFD700',       # Yellow (Gold)
    'Venstre': '#ADD8E6',   # Light Blue
    'MDG': '#008000',       # Green
    'Rodt': '#8B0000',      # Dark Red
    'Andre': '#808080'      # Gray
}

# === Use bootstrap results ===
forecast_df = mean_df
forecast_lower_df = lower_df
forecast_upper_df = upper_df

# --- Plot ---
plt.figure(figsize=(14, 7))

months_back = 12
df_recent = df.iloc[-months_back:]  # last 12 months actual

for party, color in colors.items():

    # Actual observed data
    plt.plot(
        df_recent.index,
        df_recent[party],
        marker="o",
        color=color,
        label=f"{party}"
    )

    # Forecast line (bootstrap mean)
    plt.plot(
        forecast_df.index,
        forecast_df[party],
        linestyle="dashed",
        color=color
    )

    # Connect last actual point to first forecast
    plt.plot(
        [df_recent.index[-1], forecast_df.index[0]],
        [df_recent[party].iloc[-1], forecast_df[party].iloc[0]],
        linestyle="dashed",
        color=color
    )

    # Confidence interval (bootstrap)
    plt.fill_between(
        forecast_df.index,
        forecast_lower_df[party],
        forecast_upper_df[party],
        color=color,
        alpha=0.15
    )

# --- Vertical monthly guides ---
dates = pd.date_range(start=df_recent.index[0], end=forecast_df.index[-1], freq="MS")
for date in dates:
    plt.axvline(date, color="gray", linestyle="dotted", alpha=0.3)

# --- Horizontal percentage lines ---
for percent in range(0, 45, 5):
    plt.axhline(percent, color="gray", linestyle="dotted", alpha=0.3)

# --- Final formatting ---
plt.xlim(df_recent.index[0], forecast_df.index[-1])
plt.ylim(0, 40)
plt.xlabel("Tid")
plt.ylabel("Prosent oppslutning")
plt.title("OneSixtyNine v2.0: Partienes oppslutning (12 mnd historikk + VECM-bootstrap prognose)")
plt.legend(loc="upper left", ncol=2)
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```
